```python

def clean_categories(x):
    try:
        if not isinstance(x, str) and not isinstance(x, list):
            return []

        if isinstance(x, list):
          return x

        if isinstance(x, str):
          return ast.literal_eval(x)

    except (SyntaxError, ValueError):
      print(type(x))

def clean_posts(x):
    try:
        if not isinstance(x, str) and not isinstance(x, list):
            return []

        if isinstance(x, list):
          return x

        if isinstance(x, str):
          return json.loads(x)

    except (SyntaxError, ValueError):
      print(type(x))

def clean_bios(x):
    try:
        if not isinstance(x, str) :
            return ""
        return x
    except (SyntaxError, ValueError):
      print(type(x))

def check_usernames_type(x):
  if not isinstance(x,str):
    print("im here",x)
  else :
    return x
import re
import string

def clean_hashtag_text(text):
    pattern = '[' + re.escape(string.punctuation) + '؟'+']'
    text = re.sub(pattern, " ", text)
    text = re.sub(r"http\S+", " ", text)
    text = re.sub(r"\d+", "", text)  # Remove numbers

    return text


def split_text(text):
    new_hashtag_row = []
    for hashtag in text.split():
        if any(char.isupper() for char in hashtag[2:-1]) and not hashtag.isupper():
            words = re.findall('[A-Z][^A-Z]*', hashtag)
            new_hashtag_row.extend(word.lower() for word in words)
        else:
            new_hashtag_row.append(hashtag.lower())
    return ' '.join(new_hashtag_row)

def clean_text(text):

  """
  This function cleans text data by removing special characters,
  punctuations, URLs, and converting all characters to lowercase.
  """
  pattern = '[' + re.escape(string.punctuation) + '؟'+']'
    # Remove symbols from the text
  text = re.sub(pattern, "", text)
  text = remove_emojis(text)
  text = re.sub(r"\d+", "", text)  # Remove numbers
  text = re.sub(r'\n', ' ', text)  # Replace newline characters with space
  text = re.sub(r'\S*@\S*\s?', '', text)
  # Remove URLs
  text = re.sub(r"http\S+", "", text)
  if text != "":
    """
    try:
      languages = [p.lang for p in detect_langs(text)]
      if not any(element in ['en', 'fr', 'ar'] for element in languages):
        return ''
    except:
      print('Error in detecting language for text', text)
      """
  return text.lower()

def format_number(number_str):
    if not isinstance(number_str, str):
      if isinstance(number_str,int):
        return number_str
      else:
        return 0
    number_str = number_str.replace(',','.')
    if 'K' in number_str:
        return int(float(number_str.replace('K', '')) * 1000)
    elif 'M' in number_str:
        return int(float(number_str.replace('M', '')) * 1000000)
    else:
        return int(float(number_str))
```

```python
!pip install langdetect
from langdetect import detect, detect_langs
import string
import re
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```

```python
!pip install iterative-stratification
import transformers
print(transformers.__version__)
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit
import numpy as np
from sklearn import preprocessing
from transformers import AutoTokenizer
from sklearn.utils import shuffle
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import tensorflow as tf
df['captions'] = df['posts'].apply(lambda x: ' '.join([p['caption'] for p in x if isinstance(p['caption'],str)]))
df['hashtags'] = df['hashtags'].apply(clean_hashtag_text)
df['hashtags'] = df['hashtags'].apply(split_text)
df['text'] = df['bio'] + df['captions']
print(df['bio'])
df['text'] = df['text'].apply(clean_text)
df['text'] = df['hashtags'] + df['text']
text_lengths = df['text'].apply(lambda x: len(x.split()))
print('average text length per user', text_lengths.mean())
desired_categories = ['Animals','Fashion & Beauty','Travel', 'Music', 'Food', 'Doctor', 'Education', 'Personal Blog', 'Sports', 'Acting' ]
balanced_df = df[
    df['username'].apply(lambda username: os.path.isfile(os.path.join(batches, f"{username}.pkl")))
]
balanced_df = balanced_df[balanced_df['categories'].apply(lambda x: any(cat in desired_categories for cat in x))]
balanced_df['categories'] = balanced_df['categories'].apply(lambda x: [cat for cat in x if cat in desired_categories] )

# Function to create balanced DataFrame
def create_balanced_df(df, min_samples):
    dfs = []
    used_indices = set()

    for category in desired_categories:
        # Filter rows with the current category
        category_df = df[df['categories'].apply(lambda x: category in x)]

        # Exclude rows that have already been added
        category_df = category_df[~category_df.index.isin(used_indices)]

        if len(category_df) > min_samples:
            sampled_df = category_df.sample(min_samples, replace=False)
        else:
            sampled_df = category_df

        # Add indices of the sampled rows to the used_indices set
        used_indices.update(sampled_df.index)

        dfs.append(sampled_df)

    # Concatenate the DataFrames
    balanced_df = pd.concat(dfs).reset_index(drop=True)
    return balanced_df

category_counts = balanced_df['categories'].explode().value_counts()
print(category_counts)
min_samples = min([v for k,v in category_counts.items()])
print('min_samples', min_samples)
balanced_df = create_balanced_df(balanced_df, min_samples)
category_counts = balanced_df['categories'].explode().value_counts()
print(category_counts)
```
